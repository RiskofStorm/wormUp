```python
import re
import cProfile
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
alph = list("абвгдеёжзийклмнопрстуфхцчшщъыьэюя")
cesar = list("абвгдеёжзийклмнопрстуфхцчшщъыьэюя")


def clearing_data(text):
    text = text.replace(","," ").replace(".", " ").replace(":","")

    parsed_text = text.split()
    parsed_text = [re.sub("(\d)+","",i) for i in parsed_text]
    parsed_text = [i for i in parsed_text if len(i) >0]
    return parsed_text


def move_cesar(cesar, password=0):
    for i in range(password):
        temp = cesar[0]
        cesar.pop(0)
        cesar.append(temp)
    return cesar


def decypher_cesar(parsed_text):
    full_message = list()
    for word in parsed_text:
        temp_word = list()
        for b in list(word):
            temp_word.append(alph[cesar.index(b)])
        full_message.append("".join(temp_word))
    return full_message


def fill_text(text,password):
    len_words = [len(i) for i in text if len(i)!= 0]
    
    cleared_text = list()
    new_text = list()
    
    pass_pos = 0
    pass_max = len(password)
    
    for i in range(len("".join(text))):
        if pass_pos == pass_max:
            pass_pos = 0
        new_text.append(password[pass_pos])
        pass_pos += 1
    
    for cnt in len_words:
        temp_word = list()
        for i in range(cnt):
            temp_word.append(new_text[i])
            
        cleared_text.append("".join(temp_word))
        new_text = new_text[cnt:]
    
    return cleared_text


def XOF_decypher(col_text,cypher_text):
    
    def fill_reverse_diagonal(df,true_alph):
        alph = true_alph[:]
        for i in true_alph:
            fill = pd.Series(alph, index=true_alph)
            df[i].fillna(fill, inplace=True)
            alph = move_cesar(alph, 1) 
        
        return df
    
    tabula_recta = pd.DataFrame(columns=alph, index=alph)
    tabula_recta = fill_reverse_diagonal(tabula_recta,alph)
    decrypted = list()

    for ind in range(len(col_text)):
        cyph = [i for i in cypher_text[ind]]
        col = [i for i in col_text[ind]]

        temp_word = list()
        
        for n in range(len(cyph)):
            temp_word.append(tabula_recta[tabula_recta[col[n]]== cyph[n]].index[0])

        decrypted.append("".join(temp_word))
    return " ".join(decrypted)
```
